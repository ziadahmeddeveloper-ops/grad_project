# 06 Flask API Test Demo

```

import json, requests, os
BASE_URL = 'http://127.0.0.1:5000'
payload_path = os.path.join('..', 'sample_requests', 'sample_batch_request.json') if os.path.basename(os.getcwd()) == 'notebooks' else os.path.join('sample_requests', 'sample_batch_request.json')
payload = json.load(open(payload_path, 'r', encoding='utf-8'))
```

```
requests.get(f'{BASE_URL}/health').json()
```

```
requests.post(f'{BASE_URL}/predict/batch', json=payload).json()
```
